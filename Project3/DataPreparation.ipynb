```python
#look at average_summaries and summaries
indexes = []
avg_output = []
avg_cadence = []
avg_resist = []
avg_speed = []
for k, v in dict(performance_df['average_summaries']).items():
    indexes.append(k)
    if len(v) == 0:
        avg_output.append(0)
        avg_cadence.append(0)
        avg_resist.append(0)
        avg_speed.append(0)
    else:
        for item in v:
            if item['display_name'] == 'Avg Output':
                avg_output.append(item['value'])
            elif item['display_name'] == 'Avg Cadence':
                avg_cadence.append(item['value'])
            elif item['display_name'] == 'Avg Resistance':
                avg_resist.append(item['value'])
            elif item['display_name'] == 'Avg Speed':
                avg_speed.append(item['value'])

avg_summary_df = pd.DataFrame(avg_output, indexes, columns=['avg_output'])
avg_summary_df['avg_cadence'] = avg_cadence
avg_summary_df['avg_resistance'] = avg_resist
avg_summary_df['avg_speed'] = avg_speed
avg_summary_df
        
        
```

```python
performance_df = performance_df.join(avg_summary_df).drop(columns='average_summaries')
performance_df.head()
```

```python
summaries = []
for i in list(performance_df['summaries']):
    for j in i:
        if j['display_name'] not in summaries:
            summaries.append(j['display_name'])

summaries
```

```python
metrics = []
for i in list(performance_df['metrics']):
    for j in i:
        if j['display_name'] not in metrics:
            metrics.append(j['display_name'])

metrics
```

```python
#get the total values from the summary dictionary
indexes = []
total_output = []
total_distance = []
total_calories = []
for k, v in dict(performance_df['summaries']).items():
    indexes.append(k)
    output_num = 0
    distance_num = 0
    calorie_num = 0
    for item in v:
        if item.get('display_name') == 'Total Output':
            output_num = item['value']
        elif item.get('display_name') == 'Distance':
            distance_num =  item['value']
        elif item.get('display_name') == 'Calories':
            calories_num = item['value']
    total_output.append(output_num)
    total_distance.append(distance_num)
    total_calories.append(calories_num)

total_summary_df = pd.DataFrame(total_output, indexes, columns=['total_output'])
total_summary_df['total_distance'] = total_distance
total_summary_df['total_calories'] = total_calories
total_summary_df
        
```

```python
performance_df = performance_df.join(total_summary_df).drop(columns='summaries')
performance_df.head()
```

```python
list(performance_df['metrics'])[1]
#just need max values for output, cadence, resistance, speed
#get max and average heart rate
```

```python
#get the total values from the summary dictionary
indexes = []
max_heart = []
avg_heart = []
max_output = []
max_resist = []
max_cadence = []
max_speed = []
for k, v in dict(performance_df['metrics']).items():
    indexes.append(k)
    max_heart_num = 0
    avg_heart_num = 0
    max_output_num = 0
    max_resist_num = 0
    max_cadence_num = 0
    max_speed_num = 0
    for item in v:
        if item.get('display_name') == 'Heart Rate':
            max_heart_num = item['max_value']
            avg_heart_num = item['average_value']
        elif item.get('display_name') == 'Output':
            max_output_num =  item['max_value']
        elif item.get('display_name') == 'Resistance':
            max_resist_num = item['max_value']
        elif item.get('display_name') == 'Cadence':
            max_cadence_num = item['max_value']
        elif item.get('display_name') == 'Speed':
            max_speed_num = item['max_value']
    max_heart.append(max_heart_num)
    avg_heart.append(avg_heart_num)
    max_output.append(max_output_num)
    max_resist.append(max_resist_num)
    max_cadence.append(max_cadence_num)
    max_speed.append(max_speed_num)

metric_df = pd.DataFrame(max_heart, indexes, columns=['max_heart'])
metric_df['avg_heart'] = avg_heart
metric_df['max_output'] = max_output
metric_df['max_resist'] = max_resist
metric_df['max_cadence'] = max_cadence
metric_df['max_speed'] = max_speed
metric_df.head()
```
