# Data Wrangling

```python
import requests
import json
import pandas as pd
import numpy as np
import xmltodict

peloton_username = 'PELOTON_USERNAME'
peloton_pw = 'PELOTON_PASSWORD'
```

## Getting Peloton Data

```python
# Get workouts from Peloton API
s = requests.Session()
base_url = 'https://api.onepeloton.com'
payload = {'username_or_email': peloton_username, 'password': peloton_pw}
s.post(base_url + '/auth/login', json=payload)
userID = s.get(base_url + '/api/me').json()['id']
workouts = json.loads(json.dumps(s.get(base_url + '/api/user/' + userID + '/workouts?limit=1000').json()).replace('null', '""'))
instructors = json.loads(json.dumps(s.get(base_url + '/api/instructor?limit=100').json()).replace('null', '""'))['data']
```

```python
with open('../Data/workouts.txt', 'w') as outfile:
    json.dump(workouts, outfile)

with open('../Data/instructors.txt', 'w') as outfile:
    json.dump(instructors, outfile)
```

```python
workout_df = pd.json_normalize(workouts['data'])
```

```python
workout_df.head()
```

```python
instructor_df = pd.json_normalize(instructors)
```

```python
instructor_df.head()
```

```python
#build workout summary dataframe based on workoutids
workout_ids = list(workout_df['id'])

workout_summary = []
workout_performance = []

for workout_id in workout_ids:
    summary_dict = json.loads(json.dumps(s.get(base_url + '/api/workout/' + workout_id).json()).replace('null', '""'))
    performance_dict = json.loads(json.dumps(s.get(base_url + '/api/workout/' + workout_id + '/performance_graph?every_n=1000').json()).replace('null', '""'))
    performance_dict['id']=workout_id
    workout_summary.append(summary_dict)
    workout_performance.append(performance_dict)
    
```

```python
with open('../Data/workout_summary.txt', 'w') as outfile:
    json.dump(workout_summary, outfile)
    
with open('../Data/workout_performance.txt', 'w') as outfile:
    json.dump(workout_performance, outfile)
```
